```python
# loading libraries
import pandas as pd
import yfinance as yf
import datetime as dt
import matplotlib.pyplot
import matplotlib
from sqlalchemy import create_engine
import psycopg2
```

```python
# can automate further using turtle or textinput prompting for ticker
ticker = 'SOXL'

stock_ticker = yf.download(ticker)
```

```python
# viewing last 5 entries of data. Index is set to date, goes to most recent date
stock_ticker.tail(5)
```

```python
# viewing data types
stock_ticker.dtypes
```

```python
# resetting index, adding year column, ticker symbol

# obtaining date
placeholder = stock_ticker.reset_index()
placeholder['Year'] = placeholder['Date'].astype(str)
placeholder['Year'] = placeholder['Year'].str[:4]
placeholder = placeholder.set_index('Date')

# adding ticker
placeholder['Ticker'] = ticker
stock_ticker = placeholder
stock_ticker
```

```python
stock_ticker.to_sql("Stock_Data",engine, if_exists = 'replace')
```
